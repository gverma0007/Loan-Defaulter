```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"credit_risk_dataset.csv")
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
df.head()
```

```python
df_num=df.select_dtypes(include=['int','float'])
```

```python
df_num.head()
```

```python
df_num.isnull().sum()
```

```python
df_cat=df.select_dtypes(include=['object'])
```

```python
df_cat.head()
```

```python
df_cat.isnull().sum()
```

```python
for i in df_num:
    df_num[i] = df_num[i].replace(np.nan , df_num[i].median())
```

```python
df_num.isnull().sum().sum()
```

```python
for i in df_cat:
    print(df_cat[i].value_counts())
    print('-----------------------------------------------------------')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
arr=le.fit_transform(df_cat["Education_Level"])
```

```python
df_cat['Education_Level']=pd.DataFrame(arr,columns=["Education_Level"])
```

```python
s=df_cat["Housing_Status"].value_counts()
```

```python
s
```

```python
df_cat=df_cat["Housing_Status"].map(s)
```

```python
df_new=pd.concat([df_num,df_cat],axis=1)
```

```python
df_new.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()
```

```python
mms.fit(df_new)
```

```python
df_new.describe().round(3)
```

```python
arr=mms.transform(df_new).round(2)
arr
```

```python
for i in df_new:
    df2=pd.DataFrame(arr,columns=[df_new.columns]).round(2)
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df2['Default']
y.value_counts()
```

```python
y.head()
```

```python
X=df2.drop(columns=['Default'])
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=.2 , stratify = y , random_state = 12)
```

```python
X_train.shape
```

```python
y_train.shape
y_train = y_train.astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train,y_train)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
y_pred=lr.predict(X_test)
```

```python
print('P_score', precision_score(y_test, y_pred))
```

```python
macro_precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
```

```python
macro_precision
```

```python
print('R_score', recall_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
